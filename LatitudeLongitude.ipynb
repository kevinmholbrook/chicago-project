```python
from config import API_KEY
import pandas as pd
import requests
```

```python
mexican_df=pd.read_csv("mexicanData_Final.csv")
pizza_df=pd.read_csv("pizzaData.csv")
coffee_df=pd.read_csv("coffeeData.csv")
```

```python
base_url = "https://maps.googleapis.com/maps/api/geocode/json"
new= []
for row in coffee_df.itertuples():
    new_dict={}
    address = row[2]
    print(row[1])
    params = {
        "address": address,
        "key": API_KEY
    }
    geo = requests.get(base_url, params).json()
    try:
        lat = geo["results"][0]["geometry"]["location"]["lat"]
        lng = geo["results"][0]["geometry"]["location"]["lng"]
    except:
        lat = 0
        lng = 0
    new_dict["Address"] = address
    new_dict["Latitude"] = lat
    new_dict["Longitude"] = lng
    new.append(new_dict)
new_df=pd.DataFrame(new)
new_df.head()
coffee_df_update=pd.merge(coffee_df, new_df,on='Address',how='inner')

coffee_df_update.head()
```

```python
def get_ltlng(dataframe):
    base_url = "https://maps.googleapis.com/maps/api/geocode/json"
    new= []
    for row in dataframe.itertuples():
        new_dict={}
        address = row[2]
        params = {
            "address": address,
            "key": API_KEY
        }
        geo = requests.get(base_url, params).json()
        try:
            lat = geo["results"][0]["geometry"]["location"]["lat"]
            lng = geo["results"][0]["geometry"]["location"]["lng"]
        except: 
            lat = 0
            lng = 0
        new_dict["Address"] = address
        new_dict["Latitude"] = lat
        new_dict["Longitude"] = lng
        new.append(new_dict)
    new_df=pd.DataFrame(new)
    new_df.head()
    dataframe_update=pd.merge(dataframe, new_df,on='Address',how='left')
    dataframe_update.drop_duplicates(subset="Address", keep="first", inplace=True)
    return(dataframe_update)
```

```python
pizza_update = get_ltlng(pizza_df)
mexican_update = get_ltlng(mexican_df)
```

```python
pizza_update.to_csv("pizzaUpdate.csv")
mexican_update.to_csv("mexicanUpdate.csv")
```
