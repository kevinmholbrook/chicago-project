```python
master_data = data1+data2+data3+data4+data5+data6+data7+data8+data9+data10+data11+data12+data13+data14+data15+data16+data17+data18+data19+data20+data21+data22+data23+data24+data25+data26+data27
# master_data
coffee_df = pd.DataFrame({'Name': master_data})
# pizza_df.head(10)
coffee2_df = coffee_df.drop(coffee_df.index[0])

coffee2_df.head(10)
```

```python
##################################################################################
############################### Clean the data ###################################
```

```python
##Extract the ZipCode and Telephone Number info from the coffee2_df["Name"] column 
new = coffee2_df["Name"].str.split("IL",n=1,expand = True)

# making separate "ziptel" column from new data frame 
coffee2_df["name2"]= new[0] 
coffee2_df["ziptel"]=new[1]

coffee2_df.drop(columns = ["Name"], inplace = True)

coffee2_df['Phone Number'] = coffee2_df['ziptel'].str[-12:]
# print(pizza2_df['Phone Number'])
coffee2_df['Zip Code'] = coffee2_df['ziptel'].str[:6]
# print(pizza2_df['Zip Code'])
coffee2_df.drop(columns =["ziptel"], inplace = True)

```

```python
coffee2_df.head(10)
```

```python
##Extract the City and Telephone Number info from the coffee2_df["Name"] column 
new = coffee2_df["name2"].str.split("Chicago",n=1,expand = True)

# making separate "name3" column from new data frame 
coffee2_df["name3"]= new[0] 
coffee2_df["chicago"]= new[1]

coffee2_df.drop(columns = ["name2"], inplace = True)
coffee2_df.drop(columns = ["chicago"], inplace = True)

#Choose the most important columns for the next part of the ETL Project
coffee2_df = coffee2_df[["name3","Phone Number","Zip Code"]]
coffee2_df.head(10)
```

```python
##Extract and separate "Name" and "Address" from the coffee3_df["name3"] column

# dropping null value columns to avoid errors 
coffee3_df.dropna(inplace = True) 
  
# new dataframe with split value columns 
new = coffee3_df["name3"].str.split(",", n = 1, expand = True) 
  
# making separate "Name" column from new data frame 
coffee3_df["Name"]= new[0] 
  
# making separate "Address" column from new data frame 
coffee3_df["Address"]= new[1] 

# Dropping old "name3" column
coffee3_df.drop(columns =["name3"], inplace = True) 

coffee3_df.head()
```
