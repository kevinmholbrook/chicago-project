```python
##################################################################################
############################### Clean the data ###################################
```

```python
##Extract the ZipCode and Telephone Number info from the pizza2_df["Name"] column 
new = pizza2_df["Name"].str.split("IL",n=1,expand = True)

# making separate "ziptel" column from new data frame 
pizza2_df["name2"]= new[0] 
pizza2_df["ziptel"]=new[1]

pizza2_df.drop(columns = ["Name"], inplace = True)

pizza2_df['Phone Number'] = pizza2_df['ziptel'].str[-12:]
# print(pizza2_df['Phone Number'])
pizza2_df['Zip Code'] = pizza2_df['ziptel'].str[:6]
# print(pizza2_df['Zip Code'])
pizza2_df.drop(columns =["ziptel"], inplace = True)
```

```python
pizza2_df.head(10)
```

```python
##Extract the City and Telephone Number info from the pizza2_df["Name"] column 
new = pizza2_df["name2"].str.split("Chicago",n=1,expand = True)

# making separate "name3" column from new data frame 
pizza2_df["name3"]= new[0] 
pizza2_df["chicago"]= new[1]

pizza2_df.drop(columns = ["name2"], inplace = True)
pizza2_df.drop(columns = ["chicago"], inplace = True)

#Choose the most important columns for the next part of the ETL Project
pizza2_df = pizza2_df[["name3","Phone Number","Zip Code"]]
pizza2_df.head(10)
```
